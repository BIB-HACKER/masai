You can interpret bias as the systematic error in predictions. If the bias is 5000, it suggests that, on average, your predictions are off by 5000 units from the true values. In other words, the model tends to consistently underpredict or overpredict by this amount across different instances in your dataset.

> However, it's important to note that bias alone doesn't provide a complete picture of prediction accuracy. The overall prediction error is influenced by both bias and variance. The total error can be decomposed into three parts: bias, variance, and irreducible error. The irreducible error is the inherent noise or unpredictability in the data that cannot be eliminated.

In the context of machine learning, variance refers to the amount by which a model's predictions would change if it were trained on a different dataset. It measures the model's sensitivity to the variations in the training data.

Therefore, while a bias of 5000 indicates a systematic deviation, you should also consider the variance and other aspects of model performance to get a comprehensive understanding of its predictive capability. It's common to evaluate models using metrics that capture both bias and variance, such as mean squared error (MSE) or mean absolute error (MAE).

> Variance is often assessed using evaluation metrics that quantify the model's performance on different datasets or subsets of data. common metrics used to measure variance indirectly are Root Mean Squared Error (RMSE), Mean Absolute Error (MAE) etc.

### Evaluating Model

Root Mean Squared Error (RMSE) is a commonly used metric to evaluate the performance of a model and is related to variance. RMSE is essentially the square root of the Mean Squared Error (MSE).

The formula for RMSE is:

where:
- \( n \) is the number of data points,
- \( y_i \) is the true value,
- \(y^_i\) is the predicted value.

RMSE is particularly useful because it shares the same scale as the target variable, making it easier to interpret. It penalizes larger errors more heavily than smaller ones due to the squaring operation in the MSE.

A lower RMSE indicates better model performance, meaning that the model's predictions are, on average, closer to the true values. Like MSE, RMSE is used as a measure of the difference between predicted and actual values, providing insights into the overall accuracy of the model's predictions.

MSE is calculated as the average of the squared differences between the predicted values and the actual values. A high MSE indicates that the model's predictions vary significantly from the true values across different datasets.

**Assumptions of Linear Regression:**

1. **Linearity**: The relationship between the independent variable and the dependent variable is linear. This means that the change in the dependent variable is proportional to the change in the independent variable.
2. **Independence**: The observations are independent of each other. This means that the value of one observation does not depend on the value of another observation.
3. **Homoscedasticity:**
   - **Definition:** Homoscedasticity, also known as homogeneity of variance, occurs when the variability of the residuals is constant across all levels of the independent variable(s).
   - **Visual Representation:** In a scatter plot of residuals against predicted values, homoscedasticity is evident when the spread of residuals remains roughly constant as the predicted values increase.
   - **Implication:** Homoscedasticity suggests that the assumptions of linear regression are met, and the model's predictions have consistent variability across the range of observed values.

4. **Heteroscedasticity:**
   - **Definition:** Heteroscedasticity occurs when the variability of the residuals is not constant across different levels of the independent variable(s).
   - **Visual Representation:** In a scatter plot of residuals against predicted values, heteroscedasticity is evident when the spread of residuals widens or narrows as the predicted values increase.
   - **Implication:** Heteroscedasticity violates the assumption of homogeneity of variance and can lead to inefficient estimates and biased standard errors. It may indicate that the model is missing some important explanatory variables or that the relationship between variables is not linear.

Addressing heteroscedasticity might involve transforming variables, adding relevant features, or using alternative modeling techniques that can handle heteroscedastic data appropriately.

> In summary, homoscedasticity refers to the constant variance of residuals, while heteroscedasticity refers to varying variance of residuals across different levels of the independent variable(s). It is essential to check for homoscedasticity, as violations can impact the validity of statistical inferences drawn from a regression model.
