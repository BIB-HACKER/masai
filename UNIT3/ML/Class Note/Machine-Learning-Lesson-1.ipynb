**How did we learn to recognize different types of animals, such as cats and dogs?**

As children, we saw many cats and dogs around us, and adults would tell us which ones were cats and which ones were dogs. Over time, we developed an understanding of the features and characteristics that distinguish cats from dogs, such as their size, fur, and behavior.

In machine learning, we can use a similar approach to teach a computer how to distinguish between cats and dogs. We show the computer many images of cats and dogs, and we label each image as either a cat or a dog (this constitutes our data). The computer then analyzes these images and looks for patterns and attributes that distinguish cats from dogs, such as the shape of their ears or the length of their fur.

Through repetition and exposure to many examples, the computer can learn to recognize these patterns and attributes and use them to classify new images as either cats or dogs. Of course, we still need to program the computer and use specific machine learning algorithms and techniques to train it effectively. But, like how we learned to recognize cats and dogs as children, the computer can learn to recognize these objects through exposure to many examples and repetition.

In theory, This looks all good, but How do we get machines to make decisions with data?

To know this, Let’s think how humans make decisions?

- Remember - Past situations
- Formulate - A general rule or pattern
- Predict - May happen in the future

> Let's consider an example to explain the same concept. Suppose the question is, "Will a student pass an exam?" The process to make an educated guess is as follows:

1. We remember the student's performance in past exams and assignments.
2. We formulate a hypothesis based on the student's past performance, such as "If the student has consistently performed well in exams and assignments, then they are likely to pass this exam as well."
3. We predict whether the student will pass the exam based on the hypothesis.

We may be right or wrong, but we are using our past experiences and logical reasoning to make an informed prediction. By analyzing the student's past performance, we can formulate a hypothesis that helps us predict their future performance. However, we need to be aware of the limitations of this approach, as there may be other factors that influence the student's performance, such as their health, personal circumstances, or external factors like the difficulty of the exam.

**Data vs Information:**

Is data and information same?

Data is like the raw materials that we use to create something. For example, think about building a house. The wood, bricks, and nails are like the data that we need to build the house. Similarly, in the world of computers, data refers to the facts and figures that are collected, stored, and processed by a computer.

On the other hand, information is what we get when we take that raw data and turn it into something meaningful. Going back to the example of building a house, once we use the wood, bricks, and nails to build the house, we have something that we can live in and use. Similarly, in the world of computers, information is what we get when we analyze and make sense of the data that we have collected.

To put it simply, data is like the ingredients we use to cook a meal, while information is the meal itself. We can't eat the raw ingredients, but when we cook them together, we get something that we can enjoy and make use of.

We now know that the "Previous Scores" column has the strongest correlation with "Performance Index". Let's try to find a way of estimating the value of "Performance Index" using the value of "Previous Scores". 

We'll try and "fit" a line using this points, and use the line to predict `Performance Index` for a given `Previous Scores`. A line on the X&Y coordinates has the following formula:

$y = wx + b$

The line is characterized two numbers: $w$ (called "slope") and $b$ (called "intercept"). 

### Model

In the above case, the x axis shows "Performance Index" and the y axis shows "Previous Scores". Thus, we're assuming the following relationship between the two:

`Performance Index = w(PreviousScores) + b`

```python
def estimate_performance(PreviousScores, w, b):
    return w * PreviousScores + b
```

```python
w = 50
b = 100
```

```python
Scores = df['Previous Scores']
Estimated_Performance_Indices = estimate_performance(Scores, w, b)
```

```python
plt.plot(Scores, Estimated_Performance_Indices);
plt.xlabel('Scores');
plt.ylabel('Estimated_Performance_Indices');
```
