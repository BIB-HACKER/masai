# Unsupervised Machine Learning - Clustering



> **Unsupervised machine learning** is a type of machine learning where the algorithm is not provided with labeled training data. Unlike supervised learning, where the algorithm learns from labeled examples, unsupervised learning involves finding patterns, relationships, or structures in the data without explicit guidance. This makes unsupervised learning particularly useful when dealing with unlabeled or unstructured datasets.

## Types of Unsupervised Machine Learning



1. **Clustering:**
   - Clustering is a common unsupervised learning technique that involves grouping similar data points together based on certain features or characteristics.
   - The goal is to identify natural groupings within the data, without any prior information about the categories.
   - Common clustering algorithms include K-means, hierarchical clustering, and DBSCAN.

2. **Dimensionality Reduction:**
   - Dimensionality reduction aims to reduce the number of features in a dataset while preserving its essential information.
   - Techniques like Principal Component Analysis (PCA) and t-Distributed Stochastic Neighbor Embedding (t-SNE) are often used for this purpose.

3. **Association:**
   - Association rule learning is used to identify interesting relationships or patterns in large datasets.
   - Popular algorithms for association include Apriori and Eclat, which are often applied in market basket analysis and recommendation systems.

4. **Generative Models:**
   - Generative models focus on generating new data samples that resemble the training data.
   - Examples include Generative Adversarial Networks (GANs) and Variational Autoencoders (VAEs).

### K-Means Clustering

> **K-means clustering** is a popular unsupervised machine learning algorithm used for partitioning a dataset into a set of distinct, non-overlapping groups or clusters. The goal of K-means is to group similar data points together and assign them to a common cluster, while keeping different clusters as distinct as possible. The algorithm is iterative and aims to minimize the sum of squared distances within each cluster.

Here's a step-by-step explanation of the K-means clustering algorithm:

**1. Initialization:**
   - Choose the number of clusters (K) that you want to create.
   - Randomly initialize K cluster centroids in the feature space. Each centroid represents the mean of the data points assigned to its cluster.

**2. Assignment:**
   - Assign each data point to the cluster whose centroid is closest. This is typically done using a distance metric, commonly the Euclidean distance.

**3. Update Centroids:**
   - Recalculate the centroids of the clusters by taking the mean of all data points assigned to each cluster.

**4. Repeat:**
   - Repeat the assignment and update steps until convergence. Convergence occurs when the centroids no longer change significantly, or a predefined number of iterations is reached.

**5. Output:**
   - The final output is a set of K clusters, where each data point belongs to the cluster with the nearest centroid.

#### Task 1: The Fit

**Dataset:** Iris dataset (a famous dataset containing the sepal and petal measurements of three species of Iris flowers).

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

We can check the "goodness" of the fit by looking at model.inertia_, which contains the sum of squared distances of samples to their closest cluster center. Lower the inertia, better the fit.

#### Task 2: Elbow Method for Optimal K
Choosing the right number of clusters (k) is crucial for K-means. The Elbow Method is a common technique used to select the optimal k by plotting the within-cluster sum of squares (WCSS) against the number of clusters.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**Output:** A line graph displaying the elbow curve where the x-axis represents the number of clusters and the y-axis represents the within-cluster sum of squares (WCSS). The "elbow" in the plot typically indicates the optimal number of clusters.

#### Interpretation

The code calculates and plots the Within-Cluster Sum of Squares (WCSS) for different numbers of clusters to apply the Elbow Method. This method is used to determine the optimal number of clusters for K-Means clustering. Here's the breakdown of the code and the output:

1. **WCSS Calculation:** The code initializes a list wcss to store the within-cluster sum of squares for different numbers of clusters ranging from 1 to 10. It iterates through this range, each time fitting the K-Means algorithm to the Iris dataset with the respective number of clusters and calculating the WCSS (kmeans.inertia_), which is a measure of how internally coherent the clusters are.
2. **Plotting:** It then plots these values against the number of clusters, creating a line graph.

Output Interpretation:

- The plot shows the WCSS values on the y-axis and the number of clusters (k) on the x-axis.
- There is a sharp decrease in WCSS as the number of clusters increases from 1 to 2 and then from 2 to 3. This decline in WCSS slows down after k=3, forming an "elbow" in the plot.
- The "elbow" suggests that increasing the number of clusters beyond 3 does not provide a proportional decrease in WCSS. Thus, the optimal number of clusters for this dataset is around 3, where adding more clusters doesn't significantly improve the compactness of the clusters.
- This finding aligns with the known structure of the Iris dataset, which contains three species of Iris flowers.

From the plot, one could interpret that the best choice for the number of clusters to use for the K-Means algorithm on the Iris dataset is 3, as this is the point on the graph where the WCSS starts to decrease at a slower rate, indicating diminishing returns on the explained variance within the data for each additional cluster.

**Key Points:**
- K-means is sensitive to the initial placement of centroids, and different initializations can lead to different final clusters.
- The algorithm may converge to a local minimum, so it's common to run it multiple times with different initializations and choose the best result.
- The number of clusters (K) is a critical parameter and often needs to be determined beforehand. Various methods, such as the elbow method or silhouette analysis, can help in selecting an appropriate K value.
