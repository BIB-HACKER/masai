8. **Distribution Analysis:**

```python
plt.figure(figsize=(5,3))
plt.hist(df['sepal length (cm)'], bins=20, color='green', edgecolor='black')
plt.title('sepal length')
plt.grid(linestyle='--')
plt.show()
```
