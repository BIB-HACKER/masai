```python
country_modes = netflix_df.groupby('country')['date_added'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()

def replace_date_added(row):
    country_mode_date = country_modes[country_modes['country'] == row['country']]['date_added'].iloc[0]
    return country_mode_date if pd.isna(row['date_added']) and not pd.isna(country_mode_date) else row['date_added']

netflix_df['date_added'] = netflix_df.apply(replace_date_added, axis=1)
```

5. Show top 10 directors, who gave the highest number of TV shows & Movies to Netflix?

```python
top_directors = netflix_df['director'].value_counts().nlargest(10).reset_index()
top_directors.columns = ['Director', 'Title Count']

# Create a horizontal bar plot using Plotly Express
fig = px.bar(top_directors, x='Title Count', y='Director', orientation='h',
             title='Top 10 Directors with Highest Number of Titles',
             labels={'Title Count': 'Title Count', 'Director': 'Director'},
             color='Title Count', color_continuous_scale='viridis')

# Show the plot
fig.show()
```
