## 10. Analyze the impact of seasonality on purchase patterns across different product categories. Develop visualizations to compare the average purchase amount and item popularity within each category across seasons, and investigate how seasonal trends may differ by gender. Include a minimum of 3 Visualizations such as:

1. Line Graph
2. Stacked Bar Chart
3. Heatmap





```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='Season', y='Purchase Amount (USD)', data=df, hue='Category',style='Category', markers=True)
plt.legend(loc='upper right', title='Category',bbox_to_anchor=(1.2,1))
plt.title('The impact of seasonality on purchase patterns across different product categories')
plt.xlabel('Season')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Season', y='Purchase Amount (USD)', data=df, hue='Gender',estimator=sum)
plt.legend(loc='upper right', title='Gender',bbox_to_anchor=(1.2,1))
plt.title('The impact of seasonality on purchase patterns across different product categories')
plt.xlabel('Season')
plt.ylabel('Agerage Purchase Amount (USD)')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(data=df,index='Season', column='Category',value='Purchase Amount (USD)')
plt.legend(loc='upper right', title='Gender',bbox_to_anchor=(1.2,1))
plt.title('The impact of seasonality on purchase patterns across different product categories')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```
