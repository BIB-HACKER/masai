### One Function

```python
def scrape_zomato(city):
    url = f"https://www.zomato.com/{city}"
    driver.get(url)
    driver.maximize_window()
    for k in range(20):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    restaurants = soup.find_all('div',{"class":"sc-hENMEE bVdEgD"})
    restaurant_data = {'Restaurant': [], 'Rating': [], 'Restaurant_Type': [],
                   'Price': [], 'Distance': [], 'Location': [], 'Link': []}
    for res in restaurants:
        try:
            name = res.find('h4',class_="sc-1hp8d8a-0 sc-ertOQY hArupt").text
        except:
            name = "NA"
        try:
            rating = res.find('div',class_="sc-1q7bklc-1 cILgox").text
        except:
            rating = "NA"
        try:
            res_type = res.find('p',class_="sc-1hez2tp-0 sc-iEPtyo bHxcOC").text
        except:
            res_type = "NA"
        try:
            price = res.find('p',class_="sc-1hez2tp-0 sc-iEPtyo dJHUYi").text
        except:
            price = "NA"
        try:
            distance = res.find('div',class_="min-basic-info-right").text
        except:
            distance = "NA"
        try:
            loc = res.find('p',class_="sc-1hez2tp-0 sc-gUlUPW icrStE").text
        except:
            loc = "NA"
        try:
            link = 'https://www.zomato.com' + res.find('a', class_="sc-bvCTgw dIvKTC")['href']
        except:
            link = 'NA'


        restaurant_data['Restaurant'].append(name)
        restaurant_data['Rating'].append(rating)
        restaurant_data['Restaurant_Type'].append(res_type)
        restaurant_data['Price'].append(price)
        restaurant_data['Distance'].append(distance)
        restaurant_data['Location'].append(loc)
        restaurant_data['Link'].append(link)
    df = pd.DataFrame(restaurant_data)
    return df
```
