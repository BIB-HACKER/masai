In a pivot table, the given terms are typically:

1. **Values:** This represents the numerical data that you want to aggregate. These are the actual data points you want to analyze. In the context of a pivot table, you specify which column(s) you want to use as the values to aggregate. For example, in a sales dataset, you might use the 'Revenue' column as values.

2. **Index (Rows):** These are the columns whose unique values will be used to form the index (rows) of the resulting pivot table. In other words, the unique values in these columns will become the index of the pivot table. For example, you might use the 'Date' column as the index if you want to analyze data over time.

3. **Columns:** These are the columns whose unique values will be used to form the columns of the resulting pivot table. Each unique value in these columns becomes a separate column in the pivot table. For example, you might use the 'Category' column as columns to analyze data based on different categories.

4. **Aggregation (Values - Aggfunc):** This specifies how to aggregate the numerical values for each combination of the index and columns. Common aggregation functions include 'sum', 'mean', 'count', etc. This is what you referred to as 'values' in your question.

5. **Filters (Mask):** You can use filters or masks to include or exclude certain data points from the pivot table. This is often done using the `pivot_table` function's `columns` and `index` parameters to select specific subsets of the data.

In the above example:
- 'Premium' is the numerical data (values).
- 'Category' is used as the index (rows).
- 'Region' is used as columns.
- 'mean' is the aggregation function.


## Advanced Cleaning


1. **Handling Outliers:**
   - **Identifying Outliers:**
     ```python
     df.describe()  # Descriptive statistics to identify potential outliers
     ```
   - **Handling Outliers:**
     ```python
     df = df[(df['Column'] >= lower_bound) & (df['Column'] <= upper_bound)]  # Removing outliers
     ```


2. **Handling Categorical Data:**
   - **Encoding Categorical Variables:**
     ```python
     df = pd.get_dummies(df, columns=['CategoricalColumn'], drop_first=True)  # One-hot encoding
     ```

3. **Handling Inconsistent Data:**
   - **Standardizing/Normalizing Data:**
     ```python
     df['Column'] = (df['Column'] - df['Column'].mean()) / df['Column'].std()  # Z-score normalization
     ```

4. **Handling Inconsistent Data Entry:**
   - **Mapping Values:**
     ```python
     df['Column'] = df['Column'].map({'Value1': 'MappedValue1', 'Value2': 'MappedValue2'})
     ```

5. **Handling Data Entry Errors:**
   - **Regular Expressions:**
     ```python
     df['Column'] = df['Column'].str.replace(r'\$', '')  # Removes dollar signs using regular expressions
     ```
