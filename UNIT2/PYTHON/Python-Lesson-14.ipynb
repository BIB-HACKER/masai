```python
details=[]
for element in elements:

    # Finding the title for the job posted
    try:
        title = element.find_element(By.CLASS_NAME,"JobCard_jobTitle__rbjTE").text
        #print(title)
    except:
        continue
    
    # Organisation name 
    org_name = element.find_element(By.CLASS_NAME,'EmployerProfile_employerName__8w0tV').text
    
    # Finding the Job link from the web-content
    link= element.find_element(By.CLASS_NAME,"JobCard_trackingLink__zUSOo").get_attribute('href')
    
    #location 
    location = element.find_element(By.CLASS_NAME,"JobCard_location__N_iYE").text
   
    #Salary Range
    try:
        sal = element.find_element(By.CLASS_NAME,"JobCard_salaryEstimate___m9kY").text
    except:
        sal = "NA"

    j_d= element.find_element(By.CLASS_NAME,"JobCard_jobDescriptionSnippet__HUIod").text

    details.append({
                    'Job Title': title,
                    'Organisation Name' : org_name, 
                    'Job Link' : link,
                    'Location' : location,
                    'Salary' : sal,
                    'Job Description' : j_d})
```

```python
jobs=pd.DataFrame.from_dict(details)
jobs
```

### Common ways to locate elements using Selenium

1. **By ID:**

```python
driver.find_element_by_id("job-location-1009134739550").text
```

2. **By Name:**

```python
#driver.find_element_by_name("element_name")
```

3. **By Class Name:**

```python
driver.find_element_by_class_name("JobCard_jobTitle__rbjTE").text
```

4. **By Tag Name:**

```python
driver.find_element_by_tag_name("h1").text
```

5. **By Link Text:**

```python
driver.find_element_by_link_text("Data Analyst").get_attribute('href')
```

6. **By CSS Selector:**

```python
driver.find_element_by_css_selector("#job-employer-1009098334135 > div.EmployerProfile_employerInfo__7VkqD.EmployerProfile_employerWithLogo__vpfvz > span").text
```

7. **By XPath:**

XPath is particularly powerful and flexible for locating elements. You can use absolute or relative XPath expressions. For example:

- *Absolute XPath:*

```python
driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[3]/div[2]/div[1]/div[2]/ul/li[30]/div/div/div[1]/div[2]").text
```

- *Relative XPath:*

```python
driver.find_element_by_xpath("//*[@id='left-column']/div[2]/ul/li[30]/div/div/div[1]/div[2]").text
```

8. **Click:**

```python
# Find an element by its CSS selector and click it
element_to_click = driver.find_elements(By.CLASS_NAME, "JobCard_trackingLink__zUSOo")
for i in element_to_click:
    i.click()
```

9. **Scroll:**

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver.get("https://www.glassdoor.co.in/Job/hyderabad-india-data-analyst-jobs-SRCH_IL.0,15_IC2865319_KO16,28.htm")
```

```python
while True:
    # Press the END key to scroll down
    ActionChains(driver).send_keys(Keys.END).perform()
    # Check if you have reached the bottom of the page
    if driver.execute_script("return window.innerHeight + window.scrollY >= document.body.scrollHeight"):
        break
```
