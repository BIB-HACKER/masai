12.Find the median of the above-generated 432 matrix along the columns

```python
line = np.linspace(100, 1000, 432)

matr = line.reshape(4, 108)

median = np.median(matr, axis=0)
```

```python
print(median)
```

13.Make another 2*4 matrix containing only elements from the above matrix, which are between 400 and 750.

```python
line = np.linspace(100, 1000, 432)

matrix = line.reshape(4, 108)

mask = (matrix >= 400) & (matrix <= 750)

filtered = matrix[mask].reshape(2*4, 21)
```

```python
print(filtered)
```

14.Print the sum of all the rows in the above 2*4 matrix. (Note: Kindly print only, do not overwrite the matrix.)

```python
line = np.linspace(100, 1000, 432)

matrix = line.reshape(4, 108)

mask = (matrix >= 400) & (matrix <= 750)

filter = matrix[mask].reshape(2*4, 21)
```

```python
print(np.sum(filter, axis=1))
```

15.Create a 4*5 array with 1 on the border and 0 inside.

```python
arr = np.zeros((4, 5))

arr[0, :] = 1
arr[-1, :] = 1
arr[:, 0] = 1
arr[:, -1] = 1
```

```python
print(arr)
```
