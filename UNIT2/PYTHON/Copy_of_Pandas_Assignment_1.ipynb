12.Create a new column **'Age_Category'** in the DataFrame **df** based on the following conditions: **'Young'** if Age is less than **25**, **'Middle'** if Age is **between 25 and 30** (inclusive), and **'Senior'** if Age is greater than **30**. Use the apply method.

```python
def age_category(age):
  if age<25:
    return 'Young'
  elif 25 <= age <=30:
    return 'Middle'
  else:
    return 'Senior'
```

```python
# for i in df['Age']:
#   if i<25:
#     df['Age_Category']='Young'
#   elif 25 <= i <=30:
#     df['Age_Category']='Middle'
#   else:
#     df['Age_Category']='Senior'

df['Age_Category']=df['Age'].apply(age_category)
df
```
