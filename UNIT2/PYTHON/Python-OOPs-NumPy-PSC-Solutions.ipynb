1. Create a simple class named `Rectangle` with properties `length` and `width`. Include a method `calculate_area` that returns the area of the rectangle.

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def calculate_area(self):
        area = self.length * self.width
        return area
```

```python
rectangle = Rectangle(3,4)
```

```python
rectangle.calculate_area()
```

3. Implement a class `Playlist` with a method `shuffle` that shuffles the order of songs. Include a method `play` that plays the songs in the current order.

```python
import numpy as np
class Playlist:
    def __init__(self,songs):
        self.songs = songs
    def shuffle(self):
        np.random.shuffle(self.songs)
    def play(self):
        for i in self.songs:
            print(f"Playing: {i}")
```

```python
songs_list = Playlist(["Aga Naga","Nenante","Something in the Way","DADA Song","Yeh Tune Kya Kiya"])
```

```python
songs_list.shuffle()
```

```python
songs_list.play()
```

4. Imagine you're developing a system to manage airline flights. Design a class `Flight` with private properties `flight_number`, `departure_time`, and `destination`. Include methods `delay` and `cancel` that modify the flight status.

```python
class Flight:
    def __init__(self,flight_number,departure_time,destination,status):
        self.__flight_number = flight_number
        self.__departure_time = departure_time
        self.__destination = destination
        self.status = "On time"
    def delay(self,new_dt):
        if self.__departure_time == new_dt:
            self.status = "On time"
        else:
            self.status = "Delayed"
    def cancel(self):
        self.status = "cancel"
```

```python
flight = Flight(1234,"12:30 PM IST","Vijayawada","NA")
```

```python
flight.status
```

```python
flight.delay("1:30 PM IST")
```

```python
flight.status
```

```python
flight.cancel()
```

```python
flight.status
```

5. Consider a simple scenario related to a `Book` class. Implement the `Book` class with encapsulated details about the book, such as `title`, `author`, and `publication year`. Provide methods to retrieve these details and to update the `publication year`. Additionally, create a method `(display_book_details)` to display the book's information.

```python
class Book:
    def __init__(self,title,author,publication_year):
        self.__title = title
        self.__author = author
        self.__publication_year = publication_year
    
    @property
    def get_title(self):
        return self.__title
    
    @property
    def get_author(self):
        return self.__author
    
    @property
    def get_publication_year(self):
        return self.__publication_year
    
    @get_publication_year.setter
    def set_publication_year(self,n):
        self.__publication_year = n
    
    def display_book_details(self):
        print(f"Book title: {self.__title},Book author: {self.__author},Publication year: {self.__publication_year}")
```

```python
sam_book = Book("The Guardian","Nicholas Sparks", 2003)
```

```python
sam_book.get_publication_year
```

```python
sam_book.set_publication_year = 2013
```

```python
sam_book.get_publication_year
```

```python
sam_book.display_book_details()
```
