3.Check for the **null values** in the dataset. If **null values** present in the dataset then fill those null values with apropiate values.

```python
df['Premium']=df['Premium'].fillna(df['Premium'].mode()[0])
df['Company']=df['Company'].fillna(df['Company'].mode()[0])
df.isnull().sum()
```

4.check for the **data types** of the **columns** in the dataframe and change them accordingly.

```python
df.dtypes
```

```python
df['Premium'] = df['Premium'].str.replace(',','') # Remove commas from the 'Premium' column
df['Premium'] = pd.to_numeric(df['Premium'])      # Convert the 'Premium' column to numeric
df['No Claim bonus discount'] = pd.to_numeric(df['No Claim bonus discount'], errors='coerce')
df['Company'] = df['Company'].astype('category')
df['Car Type'] = df['Car Type'].astype('category')
df['RTO'] = df['RTO'].astype('category')
df['Paymentmode'] = df['Paymentmode'].astype('category')
df['Total IDV'] = pd.to_numeric(df['Total IDV'], errors='coerce')
df['Acquisition Channel'] = df['Acquisition Channel'].astype('category')

df.dtypes
```
