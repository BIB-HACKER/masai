```python
names=[]
locations=[]
founded=[]
employees=[]
abouts=[]
skills=[]
for i in driver.find_elements(By.CLASS_NAME,'employer-job-name'):
    try:
        names.append(i.text)
    except:
        names.append('NA')
for i in driver.find_elements(By.CLASS_NAME,'employer-locations'):
    try:
        locations.append(i.text)
    except:
        locations.append('NA')
for i in driver.find_elements(By.XPATH,'//*[@id="employer-profile-opportunity"]/div[3]/span[1]/span'):
    try:
        founded.append(i.text)
    except:
        founded.append('NA')
for i in driver.find_elements(By.XPATH,'//*[@id="employer-profile-opportunity"]/div[3]/span[3]/span'):
    try:
        employees.append(i.text)
    except:
        employees.append('NA')
for i in driver.find_elements(By.XPATH,'//*[@id="employer-profile-opportunity"]/div[4]'):
    try:
        abouts.append(i.text)
    except:
        abouts.append('NA')
for i in driver.find_elements(By.CLASS_NAME,'tags.candidate-opp-keywords'):
    try:
        skills.append(i.text)
    except:
        skills.append('NA')
        
while '' in names:
    names.remove('')
    
while '' in locations:
    locations.remove('')

founded = list(filter(None, founded))

final_diction = {'Name':names,'Location':locations,'Founded':founded,'Employees':employees,'About':abouts,'Skills':skills}
    
df=pd.DataFrame.from_dict(final_diction,orient="index").T
```
