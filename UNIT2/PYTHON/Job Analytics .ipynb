```python
def download_page(url):
    response = requests.get(url, headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"})
    if response.status_code==200:
        return response.text
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None
```

```python
doc = download_page("https://www.instahyre.com/python-jobs")
```

```python
def get_name(doc):
    add_names=[]
    soup=BeautifulSoup(doc, 'html.parser')
    job_list = soup.find_all('div', class_='employer-job-name')
    
    for job in job_list:
        name = job.find('div', class_='company-name')
        add_names.append(name.text.strip())
    return add_names
```
